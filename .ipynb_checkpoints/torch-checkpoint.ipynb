## Pytorch

In this section I use pytorch in order to train neural networks for solving the problem.

```python
# =============================================================================
# Imports
# =============================================================================
import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from torch.utils.data import Dataset
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
#from training import trainNN
```

### Input features import and reshaping

```python
X = np.load(os.path.join("data", "partB", "indianpinearray.npy"))
print("Initial shape: ", X.shape)
X = X.reshape(-1,200)
print("Final shape: ", X.shape)
```

### Ground truth import and reshaping

```python
y = np.load(os.path.join("data", "partB", "IPgt.npy"))
print("Initial shape: ", y.shape)
y = y.reshape(-1)
print("Final shape: ", y.shape)
print("Classes: ", np.unique(y))
```

### Data cleaning 

There is also class 0 in the dataset. I remove class 0 which corresponds to unlabeled data.

```python
X = X[y!=0]
y = y[y!=0]
y=y-1
print("Final shapes: ", X.shape, y.shape)
print("Classes: ", np.unique(y))
```

### Dataset function to be used by the dataloader 

I implement a dataset function that its main purpose is to feed the models with batches of data. In larger datasets this function can be used to read data straight from the disk by providing list of indices  to the feats and labels variables instead of the actual features and performing the access to the disk inside the _ _ getitem _ _ method. This is more kind of a simulation of this process.

```python
class IndianPineArray(Dataset):
    
    def __init__(self, X, y):
        self.X = X
        self.y = np.array(y).astype('int64')
        
    def __getitem__(self, index):
        #features = np.load("indianpinearray")
        return {'bands': self.X[index], 'labels': self.y[index]}
    
    def __len__(self):
        return len(self.X)
```

### Train/ Test split & Normalisation

In this section I perform the splitting to training and test set. As this test set was used for hyperparameter tuning that is: early stopping and checkpoints, the definition of the architecture of the MLPs, as well as the regularization factors and the learning rates I will refer to it as "validation set" from now on.

```python
# Split to training and test set
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=69, stratify=y)

# Normalisation
sc = preprocessing.StandardScaler()
X_train = sc.fit_transform(X_train)
X_val = sc.fit_transform(X_val)

# Split training set to training and validation set
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=69, stratify=y_train)
print('Training set shape      : ', X_train.shape)
print('Training labels shape   : ', y_train.shape)
print('Validation set shape    : ', X_val.shape)
print('Validation labels shape : ', y_val.shape)
#print('Test set shape          : ', X_test.shape)
def hist(y, title):
    freq = [list(y).count(i) for i in np.unique(y)]
    print(title, freq)
    plt.bar(np.unique(y), height=freq)
    plt.xticks(np.unique(y))
    plt.show()
    return plt
hist(y_train, "Training set frequencies")
hist(y_val, "Test set frequencies")
```

The histograms highlights the obvious imbalance between the number of samples of each class

```python
train = IndianPineArray(X_train, y_train)
val = IndianPineArray(X_val, y_val)
#test = IndianPineArray(X_test, y_test)

trainloader = DataLoader(train, batch_size=128, num_workers=0)
valloader = DataLoader(val, batch_size=128, num_workers=0)
#testloader = DataLoader(test, batch_size=32, num_workers=0)
```

### Training and validation process of the models
In this section I instantiate the models, I choose loss function (Cross Entropy Loss) and the optimizer (Adam)
I also train the NNs. I also use the validation set in order to locate the appropriate number of epochs that prevents overfitting. I set the patience variable which acts as a countdown after the less validation loss is located. The model is stored in the drive as a .pt file **- checkpoint -** and if no other epoch achieves less loss the training stops after a number of epochs defined by the patience parameter. <br>
**Important:** The CrossentropyLoss function in pytorch needs to be feeded straigth with the logit outputs of the neural network in terms of prediction and with class number inside 0..15 in terms of ground truth. 

**Training and plotting function**

```python
from nn_training_evaluation import trainNN, logit_accuracy
```

**Shallow MLP**

```python
# Patience is the number of epochs that we have to wait for no
# loss decrease in the validation set to apply early stopping
# instead of waitng till the end of epochs (999)
lr = 0.0009
patience = 20
epochs = 999
regularizer = 0.000
num_classes = 16

# Instantiate model
nn1 = MLP1(input_dim=X_train.shape[1], output_dim=16).to(device)
print(nn1)
# Set loss and optimizer
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(nn1.parameters(), lr=lr, weight_decay=regularizer)
# training
_, _, _ = trainNN(model=nn1, patience=patience, best_model_name="MLP1",
                epochs=epochs, trainloader=trainloader, valloader=valloader, 
                device=device, optimizer=optimizer, criterion=criterion)
```

The shallow MLP (1 hidden layer) achieves a stable performance of around 91% in around 50 epochs using a low learning rate 0.0003 and a regularization coefficient of 0.001. Early stopping is applied by storing the model of the red dot in the disk so that overfitting is preventeed.

**Mid MLP**

```python
lr = 0.0003
patience = 40
regularizer = 0.0

# Instantiate model
nn1 = MLP2(input_dim=X_train.shape[1], output_dim=16).to(device)
print(nn1)
# Set loss and optimizer
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(nn1.parameters(), lr=lr, weight_decay=regularizer)
# training
_, _, _ = trainNN(model=nn1, patience=patience, best_model_name="MLP2",
                epochs=epochs, trainloader=trainloader, valloader=valloader, 
                device=device, optimizer=optimizer, criterion=criterion)
```

The medium-depth MLP (2 hidden layers) achieves its best best performance (~92%) at around 60 epochs. We also store the model parameters as calculated at epoch 60 for further evaluation purposes later in this excercise.

**Deep MLP**

```python
lr = 0.00009
patience = 50 # initialise here
regularizer = 0.0001

# Instantiate model
nn2 = MLP3(input_dim=X_train.shape[1], output_dim=num_classes).to(device)
print(nn2)
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(nn2.parameters(), lr=lr, weight_decay=regularizer)

# training
_, _, _ = trainNN(model=nn2, patience=patience, best_model_name="MLP3",
                  epochs=epochs, trainloader=trainloader, valloader=valloader, 
                  device=device, optimizer=optimizer, criterion=criterion)
```

The deep MLP (7 hidden layers) achieves smaller training loss and accurach much quicklier but validation accuracy does not improve -> overfit!
